```python
# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function
# to re-index the DataFrame with a new index that starts from 1 and increments by 2 for
# each row.

def reindex_dataframe(df1):
    df1= df1.reset_index(drop=True)
    new_index = pd.Series(range(1, len(df1) * 2, 2))
    df1.index = new_index
```

```python
# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
# iterates over the DataFrame and calculates the sum of the first three values in the 'Values'
# column. The function should print the sum to the console.


def calculate_sum_first_three(df5):
    # Check if the DataFrame is not empty and has at least three rows
    if not df5.empty and len(df5) >= 3:
        # Select the first three values from the 'Values' column using iloc
        first_three_values = df5['Values'].iloc[:3]
        
        # Calculate the sum of the first three values
        sum_first_three = first_three_values.sum()
        
        # Print the sum to the console
        print("Sum of the first three values:", sum_first_three)
    else:
        print("DataFrame is empty or has less than three rows.")
```

```python
# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python
# function to create a new column 'MovingAverage' that contains the moving average of the
# sales for the past 7 days for each row in the DataFrame. The moving average should be
# calculated using a window of size 7 and should include the current day.


data1 = {'Sales' : [10,5,21],
        'Date' : ['03-05-2024', '05-05-2024', '10-05-2024']}
df6 = pd.DataFrame(data1)
```

```python
def roll(df6):
    df6['MovingAverage'] = df6['Sales'].rolling(window=7, min_periods=1).mean()
    return df6
```

```python
# Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates
# a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name
# (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.


date = {"Date" : ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df7 = pd.DataFrame(date)
```

```python
# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
# function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

def select_rows_in_date_range(df):
    # Convert 'Date' column to datetime if it's not already
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Define the start and end dates of the range
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    
    # Use boolean indexing to select rows within the date range
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows
```
